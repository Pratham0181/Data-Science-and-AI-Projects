```python
# Write a Python program that checks if a given number is even. If it is, print "Even number".

n = 8
if n%2==0 :
  print("n is even")
else:
  print("n is odd")
```

```python
# Write a Python program that checks if a given string contains the letter 'a'. If it does, print "String contains 'a'"

str1 = "A"
if "a" or "A" in str1:
  print("str1 has a")
else:
  print("str1 does not have a")
```

```python
# Write a Python program that checks if a given year is a leap year. If it is, print "Leap year"

n = 2000
if (n%4==0 and n%100!=0) or n%400==0:
  print("it is a leap year")
else:
  print("not a leap year")
```
