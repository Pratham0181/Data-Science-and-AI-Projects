```python
# Create a class named Rectangle with attributes length and width. Include methods to calculate the area and perimeter of the rectangle.
# Create an instance of the Rectangle class with length 5 and width 3, then print its area and perimeter.

class Rectangle():
    def __init__(self, l, w):
        self.length = l
        self.width  = w

    def rectangle_area(self):
        return self.length*self.width

newRectangle = Rectangle(5, 3)
print(newRectangle.rectangle_area())
```
